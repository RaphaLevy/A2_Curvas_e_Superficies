## Helicoide:

```python
u, v = var('u,v', domain='real')
Helicoide = ParametrizedSurface3D((v*cos(u), v*sin(u), u), (u, v)); Helicoide
parametric_plot3d((v*cos(u), v*sin(u), u), (u,0,3*pi), (v,-pi,pi))
```

```python
u, v = var('u, v')
f = v*cos(u)
g = v*sin(u)
h = u

print('Primeira derivada')
print(' ')
print(f'Derivada de v*cos(u): {diff(f, u)}')
print(f'Derivada de v*sin(u): {diff(g, u)}')
print(f'Derivada de u: {diff(h, u)}')

print('Segunda derivada')
print(' ')
print(f'Segunda derivada de v*cos(u): {diff(f, u, 2)}')
print(f'Segunda derivada de v*sin(u): {diff(g, u, 2)}')
print(f'Segunda derivada de u: {diff(h, u, 2)}')
```

```python
p = vector([-v*sin(u), v*cos(u), 1]); q = vector([-v*cos(u), -v*sin(u), 0])
r = p.cross_product(q); r
```

OBS: p = f'; q = f''

```python
r.norm(2)
```

```python
(p.norm(2))^3
```

```python
k = r.norm(2)/((p.norm(2))^3)
k
```

## Esfera:
