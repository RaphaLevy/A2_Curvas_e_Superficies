```python
u, v = var('u, v')
f = v*cos(u)
g = v*sin(u)
h = u

print('f = v*cos(u)')
print('g = v*sin(u)')
print('h = u')

print(' ')
print('-------------------------------------------')
print(' ')

X_u = vector([diff(f, u), diff(g, u), diff(h, u)])
print(f'Primeira derivada (X_u): {X_u}')
print(' ')

X_uu = vector([diff(f, u, 2), diff(g, u, 2), diff(h, u, 2)])
print(f'Segunda derivada (X_uu): {X_uu}')


print(' ')
print('-------------------------------------------')
print(' ')

X_v = vector([diff(f, v), diff(g, v), diff(h, v)])
print(f'Primeira derivada (X_v): {X_v}')
print(' ')

X_vv = vector([diff(f, v, 2), diff(g, v, 2), diff(h, v, 2)])
print(f'Segunda derivada (X_vv): {X_vv}')

print(' ')
print('-------------------------------------------')
print(' ')

X_uv = vector([diff(f, u, v), diff(g, u, v), diff(h, u, v)])
print(f'Segunda derivada (X_uv): {X_uv}')
print(' ')

X_vu = vector([diff(f, v, u), diff(g, v, u), diff(h, v, u)])
print(f'Segunda derivada (X_vu): {X_vu}')
print(' ')
```
